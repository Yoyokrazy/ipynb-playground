```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.plot(x, y)
plt.title('Simple Sine Wave')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()

counter = 10

def increment():
    global counter
    counter += 1
    return counter

increment()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.bar(range(10), [random.randint(1, 1000) for _ in range(10)], color='blue', alpha=0.7)
plt.title('Simple Bar Chart')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)

result = fib(10)
print(f"The 10th Fibonacci number is: {result}")
```
