# Basic Syntax

Examples of basic Python syntax, including variable assignments and print statements.


```python
# Basic Syntax

# Variable assignments
x = 10  # Assigning an integer value to a variable
y = 5.5  # Assigning a float value to a variable
name = "Alice"  # Assigning a string value to a variable
is_student = True  # Assigning a boolean value to a variable

# Print statements
print(x)  # Output: 10
print(y)  # Output: 5.5
print(name)  # Output: Alice
print(is_student)  # Output: True

# Multiple assignments
a, b, c = 1, 2, 3  # Assigning multiple variables in a single line
print(a, b, c)  # Output: 1 2 3

# Basic arithmetic operations
sum_result = x + y  # Addition
print(sum_result)  # Output: 15.5

difference = x - y  # Subtraction
print(difference)  # Output: 4.5

product = x * y  # Multiplication
print(product)  # Output: 55.0

quotient = x / y  # Division
print(quotient)  # Output: 1.8181818181818181

remainder = x % y  # Modulus
print(remainder)  # Output: 4.5
```

# Data Types

Examples of different data types in Python, such as integers, floats, strings, lists, tuples, and dictionaries.


```python
# Data Types

# Integer
integer_value = 42
print(integer_value)  # Output: 42

# Float
float_value = 3.14159
print(float_value)  # Output: 3.14159

# String
string_value = "Hello, World!"
print(string_value)  # Output: Hello, World!

# List
list_value = [1, 2, 3, 4, 5]
print(list_value)  # Output: [1, 2, 3, 4, 5]

# Tuple
tuple_value = (10, 20, 30)
print(tuple_value)  # Output: (10, 20, 30)

# Dictionary
dict_value = {"name": "Alice", "age": 25, "city": "Wonderland"}
print(dict_value)  # Output: {'name': 'Alice', 'age': 25, 'city': 'Wonderland'}
```

# Control Structures

Examples of control structures in Python, including if statements, for loops, and while loops.


```python
# Control Structures

# If statements
if x > y:
    print("x is greater than y")  # Output: x is greater than y
else:
    print("x is not greater than y")

# For loops
for i in range(5):
    print(i)  # Output: 0 1 2 3 4

# Iterating over a list
for item in list_value:
    print(item)  # Output: 1 2 3 4 5

# While loops
count = 0
while count < 5:
    print(count)  # Output: 0 1 2 3 4
    count += 1
```

# Functions

Examples of defining and calling functions in Python, including functions with parameters and return values.


```python
# Functions

# Defining a simple function
def greet():
    print("Hello, World!")

# Calling the function
greet()  # Output: Hello, World!

# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

# Calling the function with an argument
greet_person("Alice")  # Output: Hello, Alice!

# Function with multiple parameters
def add_numbers(a, b):
    return a + b

# Calling the function and printing the result
result = add_numbers(10, 5)
print(result)  # Output: 15

# Function with default parameter values
def greet_person_with_default(name="Guest"):
    print(f"Hello, {name}!")

# Calling the function without an argument
greet_person_with_default()  # Output: Hello, Guest!

# Calling the function with an argument
greet_person_with_default("Bob")  # Output: Hello, Bob!

# Function with a return value
def square(number):
    return number * number

# Calling the function and printing the result
squared_value = square(4)
print(squared_value)  # Output: 16

# Function with variable number of arguments
def sum_all(*args):
    return sum(args)

# Calling the function with different number of arguments
print(sum_all(1, 2, 3))  # Output: 6
print(sum_all(4, 5, 6, 7))  # Output: 22

# Function with keyword arguments
def print_info(name, age):
    print(f"Name: {name}, Age: {age}")

# Calling the function with keyword arguments
print_info(name="Alice", age=25)  # Output: Name: Alice, Age: 25
print_info(age=30, name="Bob")  # Output: Name: Bob, Age: 30
```
