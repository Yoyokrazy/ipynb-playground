```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.plot(x, y)
plt.title('Simple Sine Wave')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()

counter = 10

def increment():
    global counter
    counter += 1
    return counter

increment()

import random
random_number = random.randint(1, 100)
print("This is a random number between 1 and 100:", random_number)
```
