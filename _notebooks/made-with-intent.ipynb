# Data Types and Variables

Show examples of different data types (int, float, string, list, tuple, dictionary) and how to create and use variables.


```python
# Data Types and Variables

# Integer
integer_var = 42
print("Integer:", integer_var)  # Output: Integer: 42

# Float
float_var = 3.14
print("Float:", float_var)  # Output: Float: 3.14

# String
string_var = "Hello, Python!"
print("String:", string_var)  # Output: String: Hello, Python!

# List
list_var = [1, 2, 3, 4, 5]
print("List:", list_var)  # Output: List: [1, 2, 3, 4, 5]

# Tuple
tuple_var = (1, 2, 3, 4, 5)
print("Tuple:", tuple_var)  # Output: Tuple: (1, 2, 3, 4, 5)

# Dictionary
dict_var = {"name": "Alice", "age": 30, "city": "New York"}
print("Dictionary:", dict_var)  # Output: Dictionary: {'name': 'Alice', 'age': 30, 'city': 'New York'}

# Using variables
x = 10
y = 20

# Addition
sum_var = x + y
print("Sum:", sum_var)  # Output: Sum: 30

# Concatenation
greeting = "Hello"
name = "Alice"
message = greeting + ", " + name + "!"
print("Message:", message)  # Output: Message: Hello, Alice!
```

# Control Flow

Provide examples of control flow statements, including if-else statements, for loops, and while loops.


```python
# Control Flow

# If-else statement
num = 10
if num > 0:
    print("The number is positive")  # Output: The number is positive
elif num == 0:
    print("The number is zero")
else:
    print("The number is negative")

# For loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
# Output:
# apple
# banana
# cherry

# While loop
count = 0
while count < 5:
    print("Count:", count)
    count += 1
# Output:
# Count: 0
# Count: 1
# Count: 2
# Count: 3
# Count: 4
```

# Functions

Define and call functions, including examples of functions with parameters and return values.


```python
# Functions

# Define a simple function
def greet():
    print("Hello from a function!")

# Call the function
greet()  # Output: Hello from a function!

# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

# Call the function with an argument
greet_person("Alice")  # Output: Hello, Alice!

# Function with return value
def add_numbers(x, y):
    return x + y

# Call the function and store the result in a variable
result = add_numbers(5, 3)
print("Sum:", result)  # Output: Sum: 8

# Function with default parameter value
def greet_with_default(name="Guest"):
    print(f"Hello, {name}!")

# Call the function without an argument
greet_with_default()  # Output: Hello, Guest!

# Call the function with an argument
greet_with_default("Bob")  # Output: Hello, Bob!
```

# Classes and Objects

Demonstrate how to define classes and create objects, including examples of class attributes and methods.


```python
# Classes and Objects

# Define a simple class
class Dog:
    # Class attribute
    species = "Canis familiaris"

    # Initializer / Instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Instance method
    def description(self):
        return f"{self.name} is {self.age} years old"

    # Another instance method
    def speak(self, sound):
        return f"{self.name} says {sound}"

# Create instances of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 5)

# Access the class attributes
print(f"{dog1.name} is a {dog1.species}")  # Output: Buddy is a Canis familiaris
print(f"{dog2.name} is a {dog2.species}")  # Output: Lucy is a Canis familiaris

# Access the instance attributes
print(dog1.description())  # Output: Buddy is 3 years old
print(dog2.description())  # Output: Lucy is 5 years old

# Call instance methods
print(dog1.speak("Woof Woof"))  # Output: Buddy says Woof Woof
print(dog2.speak("Bark"))  # Output: Lucy says Bark
```
